```python
# import statements
from time import sleep
from json import dumps
from kafka3 import KafkaProducer
import random
import pandas as pd
import datetime as dt
import json
import csv

#  home
# hostip = "192.168.1.101" # change it to your IP

# library
# hostip = '10.192.1.37'

hostip = '10.192.0.143'

def publish_message(producer_instance, topic_name, key, value):
    try:
        key_bytes = bytes(key, encoding='utf-8')
        value_bytes = bytes(value, encoding='utf-8')
        
        producer_instance.send(topic_name, key = key_bytes, value=value_bytes)
        
        producer_instance.flush()
        print('Message published successfully. Data: ' + str(value))
        
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=[f'{hostip}:9092'],
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'PartB1'
    array = []
    
    # for finding the latest date.
    with open('hotspot_AQUA_streaming.csv', 'r') as file:
        reader = csv.DictReader(file)
        
        for row in reader:  
            array.append(row)
    
    # in case, if the data should be related with producer 1 question
    # i made the latest_date as 2024,jan 1
    producer = connect_kafka_producer()
    latest_date = dt.datetime(2024, 1, 1)
    
    for _ in range(len(array)):
        random_number = random.randrange(0, len(array))
        random_data = array[random_number]
        
        # ramdomize the time
        latest_date = latest_date.replace(hour=random.randint(0, 23),
                                          minute=random.randint(0, 59),
                                          second=random.randint(0, 59))


        # set date to string 
        random_data['created_time'] = latest_date.isoformat()
        random_data['producer'] = 'producer2'
        
        #key = producer2
        publish_message(producer, topic, "producer2", json.dumps(random_data))
        sleep(3)
```
